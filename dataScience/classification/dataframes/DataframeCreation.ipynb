```
os.walk("/content/drive/MyDrive/SDGP/Dataset/Crime")
folder_path_list=[]
var = [x for x in os.walk("/content/drive/MyDrive/SDGP/Dataset/Crime")]
size=len(var)

i = 1
while i < size:
  folder_path_list.append(var[i][0])
  i+=1
print(folder_path_list)

folder_names=var[0][1]
print(folder_names)
```

```
data=[]
i=0
for folder_path in folder_path_list:
  
  os.walk(folder_path)
  var = [x for x in os.walk(folder_path)]
  file_name_list=var[0][2]
  for file_name in file_name_list:
    
    temp=[]
    file_path = folder_path+"//"+file_name
    current_file = open(file_path, "r")

    temp.append("crime")
    temp.append(folder_names[i])
    temp.append(file_name)        
    temp.append(current_file.read())

    current_file.close()
    data.append(temp)
 
  i+=1  
  

current_df = pd.DataFrame(data, columns=['category','legName','pieceName','content'])
current_df
```
